# Question 2, What city has the highest number of sales?

```python
def get_city(address):
    return address.split(",")[1]
def get_state(address):
    s=address.split(",")[2]
    return s[0:3]
def get_city_state(address):
    return get_city(address)+", "+get_state(address)
```

```python
allMonthsData["City"]=allMonthsData["Purchase Address"].apply(lambda x: get_city_state(x))
```

```python
allMonthsData.reset_index(inplace=True)
```

```python
allMonthsData.drop("index",axis=1)
```

```python
citySalesValue=allMonthsData.groupby("City").sum()["Sales Value"]
city=[city for city, df in allMonthsData.groupby("City")] 
```

```python
plt.bar(city,citySalesValue)
plt.xlabel("City Name")
plt.ylabel("Sales Value")
plt.xticks(city,rotation="vertical",size=8)
```

```python
allMonthsData.groupby("City").sum()
```

# what time should we display advertisements to maximise likelyhood of customer buying product

```python
allMonthsData.drop('index',axis=1,inplace=True)
```

```python
allMonthsData["Order Date"]=pd.to_datetime(allMonthsData["Order Date"])
```

```python
allMonthsData.head()
```

```python
allMonthsData["Hour"]=allMonthsData["Order Date"].dt.hour
allMonthsData["Minute"]=allMonthsData["Order Date"].dt.minute
allMonthsData.head()
```

```python

hour=[hour for hour, df in allMonthsData.groupby("Hour")]

plt.plot(hour,allMonthsData.groupby("Hour").count(),marker="o",markerfacecolor="red")
plt.grid()
plt.xticks(hour)
plt.xlabel("Hours in 24 hour clock")
plt.ylabel("Number of Orders")
```

```python
# 11am- 12pm  there is a peak, so we should prioritize ads during this time
# 6pm-7pm there is a peak , so we should prioritize ads during this time
```
